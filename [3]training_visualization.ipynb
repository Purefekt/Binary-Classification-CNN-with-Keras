# Visualization for Accuracy Over Time

```python
# import all libraries
import json
import matplotlib.pyplot as plt
```

```python
# Read the log json file
f = open('log_gpu.json',)
log = json.load(f)

print(log.keys())
```

## Plot Accuracy over Epochs for Training and Validation Sets

```python
training_accuracy = log['acc']
validation_accuracy = log['val_acc']
epochs = [x for x in range(1,31)]

# plot
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(epochs, training_accuracy, label = 'Training Accuracy')
plt.plot(epochs, validation_accuracy, label = 'Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# plt.savefig('accuracy.png', dpi=300, bbox_inches='tight')
plt.show()
```
